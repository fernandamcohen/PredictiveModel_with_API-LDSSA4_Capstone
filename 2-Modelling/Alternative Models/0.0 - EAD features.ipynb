```python
target1 = ['Arrest','Article found - Detailed outcome unavailable','Caution (simple or conditional)','Community resolution',
'Khat or Cannabis warning','Offender cautioned','Offender given drugs possession warning','Offender given penalty notice',
'Penalty Notice for Disorder','Summons / charged by post','Suspect arrested','Suspect summonsed to court']

target2 = ['Arrest','Article found - Detailed outcome unavailable','Summons / charged by post','Suspect arrested',
'Suspect summonsed to court']
```

```python
train['target1'] = train['Outcome'].isin(target1)
train['target2'] = train['Outcome'].isin(target2)
```

```python
print(train['target1'].value_counts(dropna=False))
print(train['target1'].value_counts(dropna=False,normalize=True))
```

```python
print(train['target2'].value_counts(dropna=False))
print(train['target2'].value_counts(dropna=False,normalize=True))
```

```python
pd.crosstab(train["Object of search"],train["target1"],normalize='index')
```

```python
pd.crosstab(train["Object of search"],train["target2"],normalize='index')
```

```python
pd.crosstab(train["station"],train["target1"],normalize='index')
```

```python
pd.crosstab(train["station"],train["target2"],normalize='index')
```
